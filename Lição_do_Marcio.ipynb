```python
# Escrever um programa que faça cálculos aritméticos

# Entrada

a = int (input("digite o primeiro número: "))
b = int (input("digite o segundo número: "))
c = int (input("digite o terceiro número: "))
d = int (input("digite a temperatura: "))
z = int (input("digite o salário: "))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b # Usar barras duplas "//" para divisões inteira
resto = a % b # O simbolo de % é igual ao resto da divisão
media = (a + b + c) / 3 # O 3 é usado para dividir o resultado da soma
fahrenheit = d * 9/5 + 32
aumento = (z / 10) + z

# Saída

print ("---------------CALCULADORA SIMPLES---------------")
print ("|                                                                     |")
print ("====================SOMA====================")
print ("|                                                                     |")
print (f"O resultado da soma de {a}  mais {b} é igual: {soma}")
print ("|                                                                     |")
print ("==================SUBTRAÇÃO=================")
print ("|                                                                     |")
print (f"O resultado da subtração de {a}  menos {b} é igual: {subtracao}")
print ("|                                                                     |")
print ("================MULTIPLICAÇÃO===============")
print ("|                                                                     |")
print (f"O resultado da multiplicação de {a} vezes {b} é igual: {multiplicacao}")
print ("|                                                                     |")
print ("===================DIVISÃO==================")
print ("|                                                                     |")
print (f"O resultado da divisão de {a} dividido {b} é igual: {divisao}")
print ("|                                                                     |")
print ("===============DIVISÃO INTEIRA==============")
print ("|                                                                     |")
print (f"O resultado da divisão de {a} dividido {b} é igual: {div_int}")
print ("|                                                                     |")
print ("====================RESTO===================")
print ("|                                                                     |")
print (f"O resultado da divisão de {a} dividido {b} o resto é igual: {resto}")
print ("|                                                                     |")
print ("====================MÉDIA===================")
print ("|                                                                     |")
print (f"O resultado da média de {a} mais {b} mais {c} dividido 3 é igual: {media}")
print ("|                                                                     |")
print ("==================CONVERÇÃO=================")
print ("|                                                                     |")
print (f"O resultado da conversão de {d} vezes por 1,8 mais 32 é igual: {fahrenheit} ")
print ("|                                                                     |")
print ("===================AUMENTO==================")
print ("|                                                                     |")
print (f"O resultado do aumento de {z} dividido por 10 mais resultado é igual: {aumento}")
print ("|                                                                     |")
print ("§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§§|")
```
